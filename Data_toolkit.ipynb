```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import bokeh.plotting as bkp
from bokeh.models import ColumnDataSource

# 1. What is NumPy, and why is it widely used in Python?
"""
NumPy (Numerical Python) is a powerful library for numerical computing in Python.
It provides high-performance multidimensional arrays and operations for mathematical
and statistical computations, making it essential for data science and machine learning.
"""

# 2. How does broadcasting work in NumPy?
"""
Broadcasting allows NumPy to perform element-wise operations on arrays of different shapes.
Smaller arrays are automatically expanded to match the larger array’s shape.
"""
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([1, 2, 3])
result = A + B  # B is broadcasted across rows

# 3. What is a Pandas DataFrame?
"""
A Pandas DataFrame is a 2D labeled data structure, similar to a table with rows and columns.
It is widely used for data manipulation and analysis.
"""
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# 4. Explain the use of the groupby() method in Pandas
"""
groupby() is used for grouping data based on a column value.
It allows for aggregation functions such as sum, mean, count, etc.
"""
df_grouped = df.groupby('A').sum()

# 5. Why is Seaborn preferred for statistical visualizations?
"""
Seaborn provides aesthetically pleasing and easy-to-use statistical visualizations.
It integrates well with Pandas and provides built-in themes and statistical plots.
"""
sns.boxplot(x=[1, 2, 3, 4, 5])
plt.show()

# 6. Differences between NumPy arrays and Python lists
"""
- NumPy arrays are more efficient for numerical operations.
- They support element-wise operations, whereas lists require loops.
- NumPy arrays use less memory.
"""

# 7. What is a heatmap, and when should it be used?
"""
A heatmap is a graphical representation of data using colors to indicate values.
It is useful for visualizing correlations, missing values, and feature relationships.
"""
data = np.random.rand(10, 10)
sns.heatmap(data, annot=True)
plt.show()

# 8. What does “vectorized operation” mean in NumPy?
"""
Vectorized operations apply computations to entire arrays without explicit loops,
leading to faster execution.
"""
array = np.array([1, 2, 3, 4])
vectorized_result = array * 2  # Element-wise multiplication

# 9. How does Matplotlib differ from Plotly?
"""
Matplotlib is static and used for fundamental plotting, while Plotly provides
interactive, web-based visualizations.
"""

# 10. What is the significance of hierarchical indexing in Pandas?
"""
Hierarchical indexing allows multiple index levels, making it useful for
multi-dimensional data representation.
"""

data = pd.Series(np.random.rand(6), index=[['A', 'A', 'B', 'B', 'C', 'C'], [1, 2, 1, 2, 1, 2]])

# 11. What is the role of Seaborn’s pairplot() function?
"""
pairplot() creates pairwise scatterplots for numerical columns in a dataset.
"""
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()

# 12. What is the purpose of the describe() function in Pandas?
"""
Provides summary statistics of a DataFrame, including mean, standard deviation, and quartiles.
"""
df.describe()

# 13. Why is handling missing data important in Pandas?
"""
Missing data can lead to inaccurate analysis; handling it ensures data integrity.
"""
df.fillna(0)  # Example of handling missing values

# 14. What are the benefits of using Plotly for data visualization?
"""
- Interactive plots
- Web-based visualization
- Easy integration with Pandas and Dash
"""

# 15. How does NumPy handle multidimensional arrays?
"""
NumPy uses n-dimensional arrays (ndarrays) for efficient multi-dimensional data handling.
"""
multi_array = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

# 16. What is the role of Bokeh in data visualization?
"""
Bokeh is a Python library for interactive, web-based visualizations.
"""

# 17. Difference between apply() and map() in Pandas
"""
- apply() is used on DataFrames and Series for applying functions row-wise or column-wise.
- map() is used for element-wise transformation on Series.
"""
df['A'].apply(lambda x: x * 2)
df['A'].map(lambda x: x * 2)

# 18. What are some advanced features of NumPy?
"""
- Broadcasting
- Vectorized operations
- Fast Fourier Transform (FFT)
- Linear algebra functions
"""

# 19. How does Pandas simplify time series analysis?
"""
Pandas provides datetime indexing, resampling, and rolling window operations for time-series data.
"""

df['date'] = pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03'])
df.set_index('date', inplace=True)
df.resample('D').sum()

# 20. What is the role of a pivot table in Pandas?
"""
Pivot tables summarize data by grouping and aggregating values.
"""
pivot = df.pivot_table(values='A', index='B', aggfunc='sum')

# 21. Why is NumPy’s array slicing faster than Python’s list slicing?
"""
NumPy arrays use contiguous memory blocks, enabling faster slicing and operations.
"""

# 22. Common use cases for Seaborn
"""
- Correlation heatmaps
- Categorical plots
- Pair plots
- Statistical analysis
"""
```

```python
# 4 A How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap)

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate random dataset
np.random.seed(42)
data = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])

# Compute correlation matrix
corr_matrix = data.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
